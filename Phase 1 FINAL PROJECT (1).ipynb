```python
# First, examine unique values in some key columns to identify issues for preprocessing
unique_summary = {
    "race": df["race"].unique(),
    "gender": df["gender"].unique(),
    "age": df["age"].unique(),
    "weight": df["weight"].unique(),
    "payer_code": df["payer_code"].unique(),
    "medical_specialty": df["medical_specialty"].unique(),
    "readmitted": df["readmitted"].unique()
}

unique_summary
```

```python
# Step 1: Drop invalid gender entries
df_cleaned = df[df["gender"] != "Unknown/Invalid"]
```

```python
# Step 2: Drop sparse/irrelevant columns
df_cleaned = df_cleaned.drop(columns=["weight", "payer_code"])
```

```python
# Step 3: Replace '?' with 'Unknown' in race and medical_specialty
df_cleaned["race"] = df_cleaned["race"].replace("?", "Unknown")
df_cleaned["medical_specialty"] = df_cleaned["medical_specialty"].replace("?", "Unknown")
```

```python
# Step 4: Encode the target variable 'readmitted'
df_cleaned["readmitted_binary"] = df_cleaned["readmitted"].apply(lambda x: 1 if x == "<30" else 0)
```

```python
# Summary of cleaned dataset
cleaned_summary = {
    "Remaining Rows": len(df_cleaned),
    "Remaining Columns": df_cleaned.shape[1],
    "readmitted_binary Counts": df_cleaned["readmitted_binary"].value_counts(),
    "race Unique Values": df_cleaned["race"].unique(),
    "gender Unique Values": df_cleaned["gender"].unique()
}

cleaned_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="darkgrid")
```

```python
# 1. Readmission distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df_cleaned, x="readmitted_binary", hue="readmitted_binary", palette="pastel", legend=False)
plt.title("Readmission Within 30 Days (Target Variable)")
plt.xticks([0, 1], ['No (<30)', 'Yes (<30)'])
plt.xlabel("Readmitted")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# 2. Gender vs Readmission
plt.figure(figsize=(6, 4))
sns.countplot(data=df_cleaned, x="gender", hue="readmitted_binary", palette="pastel")
plt.title("Gender vs Readmission")
plt.xticks(rotation=0)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Readmitted", labels=["No", "Yes"])
plt.tight_layout()
plt.show()
```

```python
# 3. Race vs Readmission
plt.figure(figsize=(8, 5))
sns.countplot(data=df_cleaned, x="race", hue="readmitted_binary", palette="pastel")
plt.title("Race vs Readmission")
plt.xticks(rotation=45)
plt.xlabel("Race")
plt.ylabel("Count")
plt.legend(title="Readmitted", labels=["No", "Yes"])
plt.tight_layout()
plt.show()
```

```python
# 4. Age distribution vs Readmission
plt.figure(figsize=(8, 5))
sns.countplot(data=df_cleaned, x="age", hue="readmitted_binary", order=sorted(df_cleaned['age'].unique()), palette="pastel")
plt.title("Age Group vs Readmission")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.legend(title="Readmitted", labels=["No", "Yes"])
plt.tight_layout()
plt.show()
```
